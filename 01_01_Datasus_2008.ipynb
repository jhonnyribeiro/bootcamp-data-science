```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = data.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
#plt.title("Valor em milhões por unidade da federação no ano/mês")
```
