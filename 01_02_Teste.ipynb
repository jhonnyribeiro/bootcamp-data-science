```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis
```

```python
usaveis= usaveis.set_index("Unidade da Federação")
usaveis
```

```python
usaveis = usaveis.sort_values("Total", ascending=False)
usaveis = usaveis.drop("Total", axis=1)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = usaveis.T.plot(figsize=(24,12))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.title("Internações por COVID-19 por UF - Região Sudeste")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = usaveis.T.plot(figsize=(24,12), kind="bar")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.title("Internações por COVID-19 por UF - Região Sudeste")
plt.grid(axis="y")
plt.xlabel('Linha temporal desde o início das internações por COVID-19')
plt.ylabel('Númmeros de internações (Milhar)')
#plt.yticks(range(1, 60000))
#plt.yticks(rotation=45)
#plt.style.use("ggplot")
plt.show()
```
